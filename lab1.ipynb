# loc[] and iloc[] in Pandas
The primary difference lies in the method of indexing data. loc[] is label-based, meaning it accesses data based on row and coloumn labels, while iloc[] is integer-based, accessing data based on integer positions.

```python
technologies={
    'Courses':["Spark","PySpark","Hadoop","Python","Pandas"],
    'Fee':[20000,25000,26000,22000,24000],
    'Duration':['30days','40days','35days','40days','60days'],
    'Discount':[1000,2300,1200,2500,2000]
}
index_labels=['r1','r2','r3','r4','r5']
df=pd.DataFrame(technologies,index=index_labels)
print(df)
```

```python
print(df.loc['r2'])
```

```python
print(df.iloc[1])
```

```python
print(df.loc[df['Fee']>=24000])
print(df.iloc[list(df['Fee']>=24000)])
```
